```python
import functions as f
from similarity.DataFrameMetadataCreator import DataFrameMetadataCreator
from collections import defaultdict
from similarity.DataFrameMetadata import DataFrameMetadata
```

```python
database, names = f.load__csv_files_from_folder("../data")
database_validation, names_validation = f.load__csv_files_from_folder("../data_validation")
```

```python
metadata: dict[str, DataFrameMetadata] = defaultdict()
for dataframe, name in zip(database, names):
    metadata[name] = DataFrameMetadataCreator(dataframe).compute_advanced_structural_types().compute_column_kind().get_metadata()
    
metadata_validation: dict[str, DataFrameMetadata] = defaultdict()
for dataframe, name in zip(database_validation, names_validation):
    metadata_validation[name] = DataFrameMetadataCreator(dataframe).compute_advanced_structural_types().compute_column_kind().get_metadata()
```

```python
first = "<class 'similarity.Types."
second = "'>"
file = open("../data/DataShow.md", "w")
for name in names_validation:
    file.write("## " + name+ "\n\n")
    print(f"## {name}\n")
    met = metadata_validation[name]
    for column in met.column_names:
        text = ('#### '+ column +': '
                '\n\t Type = '+ str(met.get_column_type(column)).replace(first, "").replace(second, "")+
                '\n\t Incomplete = '+ str(met.column_incomplete[column])+
                '\n\t Kind = '+ str(met.get_column_kind(column))+ '\n')
        file.write(text)
        print(text)
        if column in met.kind_metadata and met.kind_metadata[column] is not None:
            file.write("\t "+str(met.kind_metadata[column])+"\n")
            print(f"\t {met.kind_metadata[column]}")  
        if column in met.nonnumerical_metadata and met.nonnumerical_metadata[column] is not None:
            # file.write("\t "+str(met.nonnumerical_metadata[column])+"\n")
            print(f"\t {met.nonnumerical_metadata[column]}")
        if column in met.numerical_metadata and met.numerical_metadata[column] is not None:
            file.write("\t "+ str(met.numerical_metadata[column])+"\n")
            print(f"\t {met.numerical_metadata[column]}")
               
file.close()

```
